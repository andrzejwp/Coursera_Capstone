```python
from bs4 import BeautifulSoup
import pandas as pd
```

## Pull data from wikipedia

1. Pull the data from wikipedia page
2. Find the ``<table class="wikitable">`` element
3. For each row, except 1st (header), pull 3 ``<td>`` elements and interpret them as code, borough, neighborhood
4. Add triple to dataframe
    

```python
import requests

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

r  = requests.get(url)

data = r.text
soup = BeautifulSoup(data)
row_count = 0

rows = []

for table in soup.find_all('table', attrs={'class': 'wikitable'}):
    for row in table.find_all('tr'):
        row_count+=1
        if(row_count > 1):
            cells = row.find_all('td')
            if len(cells)==3:
                code,borough,nbhood = cells
                if(borough.getText() != 'Not assigned'):
                    rows.append([code.getText(), borough.getText(), nbhood.getText().replace('\n','')])
            #for code,borough,nbhood in row.find_all('td'):
            #    print(code)
df = pd.DataFrame(rows, columns=['PostalCode', 'Borough', 'Neighborhood'])
df.head()
```

```python
grouped = df.groupby(['PostalCode','Borough'])['Neighborhood'].apply(','.join).reset_index()
grouped.head(n=15)
```
